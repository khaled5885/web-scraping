```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import csv
```

```python
#p =input("please enter a page 1:4 ")?page={p}
url='https://www.rottentomatoes.com/browse/movies_at_home/sort:popular'
```

```python
r=requests.get(url)
r
```

```python
soup=bs(r.text,"lxml")
soup
```

```python
table=soup.find_all('div',class_='discovery-tiles__wrap')
table
```

```python
column_data = soup.find_all('a', {'data-qa':'discovery-media-list-item-caption'})
column_data
```

```python
Mdate=soup.find_all('span',class_='smaller')

table_date = [date.text.strip().replace('Streaming ', '').split(', ')[0] for date in Mdate]
table_date
```

```python
Mtitle=column_data[0].find_all('span',{'class':'p--small'})
#titles = [title.text.strip() for title in Mtitle]
Mtitle
```

```python
MTscore=soup.find_all('score-pairs-deprecated')
MTscore
```

```python
audience_score = MTscore[0].get('audiencescore')
audience_score
```

```python
critic_score = MTscore[0].get('criticsscore')
critic_score
```

```python
f=open('test/rottentomatoes.csv','w')
header='Date; Title; Audiencescore; Criticscore\n'
f.write(header)
```

```python
for M in column_data [0:]:
    Mdate=M.findAll('span',class_='smaller')[0]
    table_date = [date.text.strip().replace('Streaming ', '').split(', ')[0] for date in Mdate]
    Date =table_date[0]
    print(Date)
    Mtitle=M.findAll('span',class_='p--small')
    titles = [title.text.strip() for title in Mtitle]
    Title=titles[0]
    print(Title)
    MTscore=M.findAll('score-pairs-deprecated')
    for score in MTscore:
        audience_score = score.get('audiencescore')
        Audiencescore=audience_score
        critic_score = score.get('criticsscore')
        Criticscore=critic_score
        print(Audiencescore)
        print(Criticscore)
    f.write(Date + ";" + Title + ";" + Audiencescore + ";" + Criticscore + "\n")
f.close()
```
