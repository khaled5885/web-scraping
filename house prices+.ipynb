```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import csv
```

```python
url='https://www.huduser.gov/periodicals/ushmc/winter2001/histdat08.htm'
```

```python
r=requests.get(url)
r
```

```python
soup=bs(r.text,"lxml")
soup
```

```python
table=soup.find_all('table')[1]
table
```

```python
titles = table.find_all('th')[2:10:]
titles
```

```python
table_titles = [title.text.strip() for title in titles]

print(table_titles)
```

```python
column_data = table.find_all('tr',{'valign':'bottom'})[1:39:]
column_data
#len(column_data)
```

```python
df = pd.DataFrame(columns = table_titles)

df
```

```python
date_data = column_data[0].find_all('th',{'align':'center'})
date_data
```

```python
for row in column_data:
    row_data = row.find_all('th')
    row_data += row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
    length = len(df)
    df.loc[length] = individual_row_data
```
