```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
url='https://www.investing.com/commodities/real-time-futures'
```

```python
r=requests.get(url)
r
```

```python
soup=bs(r.text,"lxml")
soup
```

```python
t=soup.find_all('div',class_='mb-6')
t
```

```python
table = soup.find_all('div',{'class':'relative dynamic-table-v2_dynamic-table-wrapper__fBEvo'})[0]
table
```

```python
titles = table.find_all('th',{'class':'datatable-v2_cell__IwP1U datatable-v2_cell--align-start__beGpn dynamic-table-v2_col-name__Xhsxv dynamic-table-v2_col-name-h__gix_M dynamic-table-v2_header-cell__vhndq !py-2'})
titles += table.find_all('th',{'class':'datatable-v2_cell__IwP1U dynamic-table-v2_header-cell__vhndq'})
titles += table.find_all('th',{'class':'datatable-v2_cell__IwP1U dynamic-table-v2_header-cell__vhndq text-right rtl:text-right'})
titles
```

```python
table_titles = [title.text.strip() for title in titles]

print(table_titles)
```

```python
column_data = table.find_all('tr')
column_data
```

```python
df = pd.DataFrame(columns = table_titles)

df
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data][1:]
    print(individual_row_data)
    length = len(df)
    df.loc[length] = individual_row_data
```
